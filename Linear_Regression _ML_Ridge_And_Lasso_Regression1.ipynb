```python
import pandas as pd

df = pd.read_csv(r"E:\Monali\Class\machineLearning\ML\17Apr\houseprice.csv")
df.head()
```

```python
#Separating dependent & independant variable

X = df.drop('price', axis=1)
y = df['price']

print('Shape of X = ', X.shape)

print('Shape of y = ', y.shape)
```

```python
#Splitting train & test sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=51)

print('Shape of X_train = ', X_train.shape)

print('Shape of y_train = ', y_train.shape)

print('Shape of X_test = ', X_test.shape)

print('Shape of y_test = ', y_test.shape)
```

```python
# applying Fearure scaling to the data

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

sc.fit(X_train)

X_train = sc.transform(X_train)

X_test = sc.transform(X_test)
```

```python
# Train Machine Learning Model

from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
#predicted values
lr.predict(X_test)
```

```python
for i in lr.predict(X_test):
    print(i)
```

```python
y_test
```

```python
# Accuracy

lr.score(X_test, y_test)
```

# using lasso & Ridge regression

```python
from sklearn.linear_model import Ridge, Lasso

rd = Ridge()

rd.fit(X_train,y_train)

rd.score(X_test, y_test)
```

```python
ls = Lasso()

ls.fit(X_train,y_train)
ls.score(X_test, y_test)
```

```python
# setting alpha =2 in ridge
rd2 = Ridge(alpha = 2)

rd2.fit(X_train,y_train)

rd2.score(X_test, y_test)
```

```python
ls2 = Lasso(alpha=2)

ls2.fit(X_train,y_train)

ls2.score(X_test, y_test)
```

```python
ls3 = Lasso(alpha = 50)

ls3.fit(X_train,y_train)

ls3.score(X_test, y_test)
```

```python
#Root Mean Squre Error

y_pred = lr.predict(X_test)

y_pred
```

```python
# mean Square error

from sklearn.metrics import mean_squared_error 
import numpy as np

mse = mean_squared_error(y_test, y_pred)

rmse = np.sqrt(mse)

print('MSE = ', mse)

print('RMSE = ', rmse)
```
